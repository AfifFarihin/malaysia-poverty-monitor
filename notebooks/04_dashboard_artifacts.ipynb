# 04 Dashboard Artifacts

Define and verify the state-only Dash dashboard artifact contract.

The stakeholder Dash dashboard is scoped to Malaysia state-level monitoring only: Overview, State Review, and Reliability Check. This notebook verifies the dashboard artifact contract after the raw DOSM lineage check has already been recorded by Notebook 00. The generated file `data/state_model_predictions.csv` is treated as a selected-model prediction artifact, not raw data.


```python
from __future__ import annotations

import json
import os
import sys
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "config").exists():
    PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))

QA_DIR = PROJECT_ROOT / "outputs" / "qa"
REPORTING_DIR = PROJECT_ROOT / "outputs" / "reporting"
INTERMEDIATE_DIR = PROJECT_ROOT / "outputs" / "intermediate"
for path in [QA_DIR, REPORTING_DIR, INTERMEDIATE_DIR]:
    path.mkdir(parents=True, exist_ok=True)

def write_json(path: Path, payload: dict) -> dict:
    path.parent.mkdir(parents=True, exist_ok=True)
    path.write_text(json.dumps(payload, indent=2, default=str), encoding="utf-8")
    return payload

def read_json(path: Path) -> dict:
    return json.loads(path.read_text(encoding="utf-8")) if path.exists() else {}
```

```python
required = [
    "data/state_year_panel_modelready_2002_2022.csv",
    "data/state_model_predictions.csv",
    "outputs/qa/dosm_raw_lineage_validation.json",
    "data/boundaries/malaysia_adm1_geoboundaries.geojson",
    "outputs/metrics/malaysia_state_selected_model_loso_metrics.json",
    "outputs/metrics/malaysia_state_model_comparison.csv",
    "outputs/metrics/malaysia_state_loso_fold_predictions.csv",
    "outputs/metrics/malaysia_state_uncertainty_intervals.csv",
    "outputs/metrics/malaysia_state_temporal_backtest.csv",
    "outputs/qa/malaysia_state_temporal_backtest.json",
    "outputs/figures/mys_state_rf_shap_poverty_absolute.png",
    "outputs/metrics/malaysia_state_rf_shap_poverty_absolute.csv",
    "outputs/qa/malaysia_state_rf_shap_summary.json",
]
contract = []
for rel in required:
    path = PROJECT_ROOT / rel
    contract.append({"path": rel, "exists": path.exists(), "size_mb": round(path.stat().st_size / 1_048_576, 3) if path.exists() else None})

summary = {
    "dashboard_scope": "Malaysia state-level stakeholder view only",
    "tabs": ["Overview", "State Review", "Reliability Check"],
    "runtime_rule": "Dash reads certified final artifacts only and does not recalculate model results.",
    "explainability_rule": "RF SHAP describes the selected sensor-safe absolute-poverty model; it is not causal evidence.",
    "artifacts": contract,
    "all_present": all(item["exists"] for item in contract),
}
write_json(QA_DIR / "dashboard_artifact_contract.json", summary)
pd.DataFrame(contract)
```
